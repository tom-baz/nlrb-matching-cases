```python
def standardize_case_number(case_number):
    """Ensure consistent format: XX-YY-ZZZZZZ (region-type-docket)"""
    parts = case_number.split('-')
    parts[0] = parts[0].zfill(2)  # Region: 2 digits
    parts[-1] = parts[-1].zfill(6)  # Docket: 6 digits
    return '-'.join(parts)


# Apply standardization to all dataframes
df_CHIPS_master_merged['case_number'] = df_CHIPS_master_merged['case_number'].apply(standardize_case_number)
chips_elections['case_number'] = chips_elections['case_number'].apply(standardize_case_number)
cases_city_zip_chips['case_number'] = cases_city_zip_chips['case_number'].apply(standardize_case_number)
cases_names_chips['case_number'] = cases_names_chips['case_number'].apply(standardize_case_number)

print("✓ Case numbers standardized")
```

## Same-employer check (C vs R employer names)

For each example, the **C-case employer** (from `cases_names_chips`) shown next to each **matched R-case employer**, so the same-establishment match is visible at a glance.

The R sets below are the petitions each charge **actually matched** in `rc_ac_matches.parquet`:
- **Example 1 corrected** — `30-CA-014400` (FELTNER PLUMBING) matched `30-RC-006036` / `30-RC-006037`, *not* the `01-RC-...` Rhode Island Hospital cases listed earlier (different employer, state, and region).
- **Example 2** includes `12-RC-007164` (the 6th petition) that was dropped from the list above.

```python
# ---- Same-employer check: C employer next to each matched R employer ----
# Uses the standardized cases_names_chips (Employer per case_number) loaded above.
names = cases_names_chips.set_index('case_number')['Employer']

# C case -> the R petitions it ACTUALLY matched in rc_ac_matches.parquet
examples = {
    '30-CA-014400': ['30-RC-006036', '30-RC-006037'],
    '12-CA-013560': ['12-RC-007164', '12-RC-007166', '12-RC-007176',
                     '12-RC-007177', '12-RC-007182', '12-RC-007183'],
    '31-CA-016742': ['31-RC-006212', '31-RC-006213', '31-RC-006214',
                     '31-RC-006215', '31-RC-006216', '31-RC-006217'],
    '19-CA-016244': ['19-RC-010501', '19-RC-010502', '19-RC-010504',
                     '19-RC-010506', '19-RC-010507'],
}

def employer_comparison(examples):
    """Tidy table: C-case employer beside each matched R-case employer, per example."""
    rows = []
    for ex, (c_case, r_cases) in enumerate(examples.items(), 1):
        rows.append({'example': ex, 'role': 'C (charge)', 'case_number': c_case,
                     'Employer': names.get(c_case, '(not in cases_names_chips)')})
        for r in r_cases:
            rows.append({'example': ex, 'role': 'R (petition)', 'case_number': r,
                         'Employer': names.get(r, '(not in cases_names_chips)')})
    return pd.DataFrame(rows)

employer_comparison(examples)
```
